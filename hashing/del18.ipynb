```python
DEFAULT_INITIAL_CAPACITY = 1 << 5 # Same as 2^5 = 32
MAX_CAPASITY = 1 << 30  # 2^30

class Map :

    def __init__(self, capasity = DEFAULT_INITIAL_CAPACITY, max_capasity = MAX_CAPASITY) :
        self.capasity = capasity
        self.table = []

        for i in range(capasity) :
            self.table.append([])    # Initialize buckets

        self.size = 0

    def put(self, key, value) :

        if self.size == self.capasity :
            raise RuntimeError("Exceeded max capasity of hashmap")

        bucketIndex = self.get_hash(key)
        bucket = self.table[bucketIndex]
        if len(bucket) >= 1 :
            print("Bucket with key", key, "has length", len(bucket))
        if not self.has_key_allready(key):
            bucket.append((key,value))
        else :
            raise RuntimeError("Cannot put, key already exists")

        self.size += 1

    def get(self, key) :
        bucketIndex = self.get_hash(key)
        if len(self.table[bucketIndex]) > 0 :
            bucket = self.table[bucketIndex]
            for element in bucket :
                if element[0] == key :
                    return element[1]
        
        return None

    def remove(self, key) :
        bucketIndex = self.get_hash(key)
        if len(self.table[bucketIndex]) > 0 :
            bucket = self.table[bucketIndex]
            for element in bucket :
                if element[0] == key :
                    bucket.remove(element)
                    self.size -= 1
                    break

    def has_key_allready(self, key) :
        bucketIndex = self.get_hash(key)
        if len(self.table[bucketIndex]) > 0 :
            bucket = self.table[bucketIndex]
            for element in bucket :
                if element[0] == key :
                    return True

        return False


    def supplemental_hash(self, h) :
        h ^= (h >> 20) ^ (h >> 12)
        return h ^ (h >> 7) ^ (h >> 4)

    def get_hash(self, key) :
        initial_val = hash(key)
#        sum = 0
#        for i in range(len(key)) :
#            sum += ord(key[i])
#        initial_val = sum
        index = initial_val % self.capasity
        distributed_val = self.supplemental_hash(initial_val)
        index = distributed_val % self.capasity
        print (index)
        return index
        return initial_val & (self.capasity-1)   # Mask out the lowest bits of distributed hash. 
                                                    # This will make an index in the range [0, self.capasity]


map = Map(capasity=4)
map.put("BAAA", 31)
map.put("AAAA", 29)
map.put("AAA", 29)
map.put("AB", 29)

# map.get("Anderson")
```
